# Data Cleaning

```python
# Dropping unnecessary columns
df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Converting data types
df['servesize'] = df['servesize'].astype('float64')
df['calories'] = df['calories'].astype('float64')
```

```python
df.info()
```

```python
df["menu"].unique()
```

# Data Visualization

```python
# Items count in menu lists
sns.countplot(x='menu',data=df)
plt.title('THE MENU LIST')
plt.xlabel('menu')
plt.ylabel('count')
plt.xticks(rotation=45)
```

```python
# Grouping according to the average value 
df.groupby("menu").mean(numeric_only=True)
```

```python
# Calories in menu list
plt.figure(figsize=(6,3))

sns.barplot(x = 'menu', y = 'calories', data = df)
plt.title('Calories in menu')
plt.xlabel('Menu')
plt.ylabel('Calories')
plt.xticks(rotation=45)
plt.show()
```

```python
# protiens in menu list

sns.stripplot(x='menu',y='protien',hue='menu',data=df)

plt.title('protiens in menu')
plt.xlabel('menu')
plt.ylabel('protien')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total fat in menu list
sns.barplot(x = 'menu', y = 'totalfat', data = df)

plt.title('Total fat in menu')
plt.xlabel('Menu')
plt.ylabel('Total Fat')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calories in menus according to the density ratio
(sns
 .FacetGrid(df,
              hue = "menu",
              height = 5,
              xlim = (0, 1000))
 .map(sns.kdeplot, "sodium", fill= True)
 .add_legend()
);
```
